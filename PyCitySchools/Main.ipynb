```python
import pandas as pd
import numpy as np
school_data = "Resources/schools_complete.csv"
school_data_df=pd.read_csv(school_data)
school_data_df.head()
student_data = "Resources/students_complete.csv"
student_data_df=pd.read_csv(student_data)
school_data_complete = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

# School Summary

```python
# # Calculate the total number of schools
school_count=school_data_df["School ID"].count()
total_students=school_data_df["size"].sum()
total_budget=school_data_df["budget"].sum()
average_m_score=student_data_df["math_score"].mean()
average_a_score=student_data_df["reading_score"].mean()
average_passing_rate=(student_data_df["math_score"].mean() + student_data_df["reading_score"].mean())/2
math_score_df=student_data_df["math_score"]>=70
reading_score_df=student_data_df["reading_score"]>=70
math_score1=math_score_df[math_score_df==True].count().sum()
reading_score1=reading_score_df[reading_score_df==True].count().sum()
math_score1=((math_score1)/(total_students).sum())*100
reading_score1=((reading_score1)/(total_students).sum())*100
district_summary_table=pd.DataFrame({"Total Schools":[school_count],
                                    "Total Students": [total_students],
                                    "Total Budget": [total_budget],
                                    "Average Math Score":[average_m_score],
                                    "Average Reading Score":[average_a_score],
                                     "% Passing Math":[math_score1],
                                     "% Passing Reading":[reading_score1],
                                    "% Overall Passing Rate":[average_passing_rate]})
district_summary_table
```

```python
final_df=school_data_complete_df
```

```python
passing_in_math_df = school_data_complete_df[student_data_df["math_score"] > 70]
```

```python
passing_in_reading_df = school_data_complete_df[student_data_df["reading_score"] > 70]
```

```python
percent_math_df=passing_in_math_df["school_name"].value_counts()
```

```python
percent_reading_df=passing_in_reading_df["school_name"].value_counts()
```

```python
counts_school_data_df=school_data_complete_df["school_name"].value_counts()
```

```python
students_passing_reading_by_school=(passing_in_reading_df.groupby("school_name").count()["Student ID"]/counts_school_data_df)*100
students_passing_reading_by_school_df=students_passing_reading_by_school
```

```python
students_passing_math_by_school=(passing_in_math_df.groupby("school_name").count()["Student ID"]/counts_school_data_df)*100
students_passing_math_by_school_df=students_passing_math_by_school
```

```python
ave_math=(passing_in_math_df.groupby("school_name").mean()["math_score"])
ave_reading=(passing_in_reading_df.groupby("school_name").mean()["reading_score"])
```

```python
school_types = school_data_df.set_index(["school_name"])["type"]
# per_school_counts = school_data_complete["school_name"].value_counts()
```

```python
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_student_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]/counts_school_data_df
```

```python
overall_passing=(students_passing_reading_by_school+students_passing_math_by_school_df)/2
overall_passing
```

# School Summary

```python
dictionary = {"School Budget":per_school_budget,
              "Per Student Budget": per_student_budget,
             "School Type": school_types,
             "Size": counts_school_data_df,
             "Ave Math Score": ave_math,
            "Ave Reading Score": ave_reading,
             "Pass Math": students_passing_math_by_school_df,
             "Pass Reading": students_passing_reading_by_school,
             "Overall Passing": overall_passing}
dictionary
pd.DataFrame(dictionary)
```

# Top Performing Schools (By Passing Rate)
